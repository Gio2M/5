# DESCOMPRESION Y LECTURA DEL DATASET
Los archivos proporcionados estan comprimido en formato Gzip que contiene datos en formato JSON. Los archivos comprimidos en formato Gzip (.gz) se utilizan para reducir el tamaño de los archivos y ahorrar espacio en disco. JSON es un formato de intercambio de datos ampliamente utilizado que almacena información en una estructura legible por humanos y fácilmente interpretable por máquinas.

 Primero, se descomprimen los archivos Gzip utilizando la biblioteca gzip, esto creará archivos descomprimidos ".json" en el mismo directorio.

DESCOMPRESION

```python
import gzip

input_file = 'Data/Datos_fuente/Datos_comprimidos/user_reviews.json.gz'
output_file = 'Data/Datos_fuente/Datos_json/user_reviews.json'

with gzip.open(input_file, 'rb') as f_in:
    with open(output_file, 'wb') as f_out:
        f_out.write(f_in.read())
```

# Feature Engineering

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Crear un DataFrame de ejemplo
#data = {'review_column': ["Este producto es terrible.", "Esto es más o menos.", "¡Amo este producto!"]}
#expanded_df_reviews = pd.DataFrame(data)

# Inicializar VADER
sid = SentimentIntensityAnalyzer()

# Función para categorizar el sentimiento
def categorize_sentiment(compound_score):
    if compound_score >= 0.05:
        return 2  # Positivo
    elif compound_score > -0.05 and compound_score < 0.05:
        return 1  # Neutral
    else:
        return 0  # Malo

# Hacer una copia explícita del DataFrame filtered_df
reviews_copy = reviews.copy()

# Calcular el puntaje de sentimiento compuesto y categorizar
reviews_copy['sentiment_score'] = reviews_copy['review'].apply(lambda x: sid.polarity_scores(x)['compound'])
reviews_copy['sentiment_analysis'] = reviews_copy['sentiment_score'].apply(categorize_sentiment)

print(reviews_copy)
```

```python
reviews_copy.info()
```

```python
# Obtener estadísticas descriptivas de la columna 'sentiment_score'
score_stats = reviews_copy['sentiment_score'].describe()

# Obtener el rango (diferencia entre el valor máximo y el valor mínimo)
score_range = score_stats['max'] - score_stats['min']

print("Estadísticas descriptivas de sentiment_score:")
print(score_stats)
print("Rango de valores en sentiment_score:", score_range)
```
